```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

```

```python
# Generate a small dataset
np.random.seed(0)
X = np.linspace(-5, 5, 100).reshape(-1, 1)
y = np.random.normal(0, 1, size=(100, 1))
```

```python
# Apply activation functions to the dataset
y_sigmoid = sigmoid(X)
y_tanh = tanh(X)
y_relu = relu(X)
```

```python
# Create subplots for each activation function
fig, axs = plt.subplots(1, 3, figsize=(18, 6))
axs[0].plot(X, y_sigmoid)
axs[0].set_title('Sigmoid')
axs[0].grid(True)

axs[1].plot(X, y_tanh)
axs[1].set_title('Tanh')
axs[1].grid(True)

axs[2].plot(X, y_relu)
axs[2].set_title('ReLU')
axs[2].grid(True)

```

```python
# Adjust layout
plt.tight_layout()
plt.show()
```
